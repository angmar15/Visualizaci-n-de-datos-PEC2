# AREA CHART

Para este apartado usaré unos datos comparando las distintas fuentes de energía usadas, descargados de la página web ourworld in data proporcionada en el enunciado. 

```python
file_path = 'owid-energy-data.csv'
#importo los datos descargados desde la pagina https://ourworldindata.org/energy y veo su estructura
energy_data = pd.read_csv(file_path)
energy_data.head(), energy_data.columns.tolist()
```

Estos datos los voy a usar para el grafico de area, asi que usaré todas las columnas referentes al consumo de energías renovables y nuclear , siempre filtrando para un país. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#filtro para un pais en especifico, en este caso eeuu

country = 'United States'
data_country = energy_data[energy_data['country'] == country]
```

```python
#selecciono las columnas elegidas
area_data = data_country[['year', 'solar_consumption', 'wind_consumption', 
                          'hydro_consumption', 'nuclear_consumption']].dropna()
```

```python
#desarrollo del grafico de area con los parámetros elegidos. 
plt.figure(figsize=(10, 6))
plt.stackplot(area_data['year'], area_data['solar_consumption'], area_data['wind_consumption'], 
              area_data['hydro_consumption'], area_data['nuclear_consumption'], labels=['Solar', 'Wind', 'Hydro', 'Nuclear'])
plt.legend(loc='upper left')
plt.title(f'Energy Consumption by Source in {country} (2000-2020)')
plt.xlabel('Year')
plt.ylabel('Consumption (TWh)')
plt.show()
```
